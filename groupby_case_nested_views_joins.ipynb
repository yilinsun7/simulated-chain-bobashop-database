```python
# Connect to the database 

import os
import pandas as pd
from configparser import ConfigParser
from mysql.connector import MySQLConnection
from database_utilities import make_connection, dataframe_query

conn = make_connection('boba_shop.ini')
cursor = conn.cursor()
```

```python
# table tea

cursor.execute('DROP TABLE IF EXISTS tea')
sql = ( """
        CREATE TABLE tea
        (
          tea_name     VARCHAR(45) NOT NULL,
          tea_price    FLOAT NOT NULL,
          topping      VARCHAR(45) NOT NULL,
          PRIMARY KEY (tea_name, topping)
        );
        
        """
      )

cursor.execute(sql)
```

```python
# table store

cursor.execute('DROP TABLE IF EXISTS store')
sql = ( """
        CREATE TABLE store
        (
          store_id       VARCHAR(20) NOT NULL,
          store_address  LONGTEXT NOT NULL,
          PRIMARY KEY (store_id)
        );
        
        """
      )

cursor.execute(sql)
```

```python
# table sells

cursor.execute('DROP TABLE IF EXISTS sells')
sql = ( """
        CREATE TABLE sells
        (
          tea_name       VARCHAR(45) NOT NULL,
          store_id       VARCHAR(20) NOT NULL,
          topping        VARCHAR(45) NOT NULL,
          PRIMARY KEY (tea_name, topping),
          FOREIGN KEY (tea_name, topping) REFERENCES tea(tea_name, topping),
          FOREIGN KEY (store_id) REFERENCES store(store_id)
        );
        
        """
      )

cursor.execute(sql)
```

```python
# table employee

cursor.execute('DROP TABLE IF EXISTS employee')
sql = ( """
        CREATE TABLE employee
        (
          emp_id           VARCHAR(10) NOT NULL,
          emp_name         VARCHAR(45) NOT NULL,
          street_num       INT NOT NULL,
          street_name      VARCHAR(20) NOT NULL,
          city             VARCHAR(20) NOT NULL,
          zip              INT NOT NULL,
          emp_dob          DATE NOT NULL,
          store_id         VARCHAR(20) NOT NULL,
          PRIMARY KEY (emp_id),
          FOREIGN KEY (store_id) REFERENCES store(store_id)
        );
        
        """
      )

cursor.execute(sql)
```

```python
# table customer

cursor.execute('DROP TABLE IF EXISTS customer')
sql = ( """
        CREATE TABLE customer
        (
          cus_id         INT NOT NULL,
          cus_name       VARCHAR(45) NOT NULL,
          pay_name       VARCHAR(45) NOT NULL,
          PRIMARY KEY (cus_id)
        );
        
        """
      )

cursor.execute(sql)
```

```python
# table vehicle

cursor.execute('DROP TABLE IF EXISTS vehicle')
sql = ( """
        CREATE TABLE vehicle
        (
          v_id      VARCHAR(20) NOT NULL,
          model     VARCHAR(20) NOT NULL,
          make      VARCHAR(20) NOT NULL,
          v_year    VARCHAR(10) NOT NULL,
          v_plate   VARCHAR(10) NOT NULL,
          emp_id    VARCHAR(10) NOT NULL,
          PRIMARY KEY (v_id),
          FOREIGN KEY (emp_id) REFERENCES employee(emp_id),
          UNIQUE (v_plate)
        );
        
        """
      )

cursor.execute(sql)
```

```python
# table orders

cursor.execute('DROP TABLE IF EXISTS orders')
sql = ( """
        CREATE TABLE orders
        (
          order_date    DATE NOT NULL,
          cus_id        INT NOT NULL,
          tea_name      VARCHAR(45) NOT NULL,
          topping       VARCHAR(45) NOT NULL,
          PRIMARY KEY (tea_name, topping),
          FOREIGN KEY (cus_id) REFERENCES customer(cus_id),
          FOREIGN KEY (tea_name, topping) REFERENCES tea(tea_name, topping)
        );
        
        """
      )

cursor.execute(sql)
```

```python
# table employee_phone_num

cursor.execute('DROP TABLE IF EXISTS employee_phone_num')
sql = ( """
        CREATE TABLE employee_phone_num
        (
          phone_num      VARCHAR(20) NOT NULL,
          emp_id         VARCHAR(10) NOT NULL,
          PRIMARY KEY (phone_num, emp_id),
          FOREIGN KEY (emp_id) REFERENCES employee(emp_id)
        );
        
        """
      )

cursor.execute(sql)
```

### 1. INSERT INTO SELECT with CASE

To make use of the INSERT INTO SELECT with CASE, we decided to create a new table that includes a detail about the vehicle age from an existing table 'vehicle' that includes information on when was the vehicle made. If this vehicle was made before the year 2000, we will label it as "Old Vehicle" in the 'age' column in our new table `vehicle_age`; if this vehicle was made after the year 2016, we will label it as "New Vehicle" in the 'age' column in our new table `vehicle_age`;

Creating our new table `vehicle_age`

```python
# table vehicle_age

cursor.execute('DROP TABLE IF EXISTS vehicle_age')
sql = ( """
        CREATE TABLE vehicle_age
        (
          age       VARCHAR(40) NOT NULL,
          plate     VARCHAR(40) NOT NULL,
          v_id      VARCHAR(40) NOT NULL
        );
        
        """
      )

cursor.execute(sql)
```

Insert the values into this new table based on when was the vehicle made

```python
sql = ( """
        INSERT INTO vehicle_age (age, plate, v_id)
        SELECT (
                CASE 
                WHEN vehicle.v_year < 2000 THEN 'Old Vehicle'
                ELSE 'New Vehicle'
                END) AS age,
                vehicle.v_plate AS plate,
                vehicle.v_id AS v_id 
        FROM vehicle
        """
      )

cursor.execute(sql)
```

Display the original table `vehicle`:

### 2. Aggregate functions with GROUP BY

Inside the table `tea`, some tea will have the same name but different toppings, resulting in different prices; we want to see how many tea has same names for each name; 

```python
display_df(conn,
            """
            SELECT tea_name, COUNT(*) 
            FROM tea
            GROUP BY tea_name
            """
          )
```

### 3. Aggregate functions with GROUP BY HAVING

Continuing our previous question: many tea have the same name but different toppings, but there are also some tea that is unique, having only one type of topping. We would like to see which tea has multiple different combination of toppings. 

```python
display_df(conn,
            """
            SELECT tea_name, COUNT(*) 
            FROM tea
            GROUP BY tea_name
            HAVING COUNT(*) > 1
            """
          )
```

### 4. Nested query

We want to know which drinks sell below the average tea price;

```python
display_df(conn,
            """
            SELECT tea_name, tea_price, topping 
            FROM tea
            WHERE tea_price < (SELECT AVG(tea_price)
                               FROM tea);
            """
          )
```

### 5. Left outer join

We want to know what tea were purchased by which payment method by the customers.

```python
display_df(conn,
            """
            SELECT c.cus_id, c.cus_name, c.pay_name, o.tea_name
            FROM customer c
            LEFT OUTER JOIN orders o
                ON c.cus_id = o.cus_id
            ORDER BY c.cus_id;
            """
          )
```

### 6. Right Outer Join

We want to know about the young employees in our boba_shop. Particularly the employees below age 25.

```python
display_df(conn,
            """
            SELECT e.emp_id, e.emp_name, e.emp_dob, eph.phone_num
            FROM employee_phone_num AS eph
            RIGHT OUTER JOIN employee e
                ON eph.emp_id = e.emp_id
            WHERE e.emp_dob BETWEEN '2000-01-01' AND '2020-12-31'
            ORDER by e.emp_id;
            """
          )
```

### 7. Full Outer Join

##### Since full outer join is not availabel in MYSQL, we will perform the operation using UNION.
##### We want to know which tea are sold among all the stores.

```python
display_df(conn,
            """
            SELECT st.store_id, st.store_address, sells.tea_name
            FROM store st
            LEFT OUTER JOIN sells 
            	ON st.store_id = sells.store_id
            UNION
            SELECT st.store_id, st.store_address, sells.tea_name
            FROM store st
            RIGHT OUTER JOIN sells
            	ON st.store_id = sells.store_id;
            """
          )
```

### 8. View

```python
cursor.execute("DROP VIEW IF EXISTS credit_card_payers")

sql =    ("""
            CREATE VIEW credit_card_payers AS
        	SELECT c.cus_id, c.cus_name, c.pay_name
            FROM customer c
        	WHERE c.pay_name = "credit card";
            """
          )

cursor.execute(sql)
```

### Displaying the VIEW

```python
display_df(conn,
            """
            SELECT *  FROM credit_card_payers ;
            """
          )
```

### Using the VIEW


To answer the question as to on which credit card payemnts were made and by whom, we'll make use of the view we created

```python
display_df(conn,
            """
            SELECT o.order_date, ccp.cus_id, ccp.cus_name, ccp.pay_name
            FROM credit_card_payers ccp
            JOIN orders o
                ON ccp.cus_id = o.cus_id;
            """
          )
```
